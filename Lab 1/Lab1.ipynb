```
# Question 2
#2.1
raw_data=sc.textFile("/FileStore/tables/q2.csv")
#2.2
lines = raw_data.map(lambda line: line.split(","))
#2.3
print(lines.count(),lines.take(1))
#2.4
print("Number of the distinct year:", lines.map(lambda x: (x[0], 1)).reduceByKey(lambda x, y: x+y).count())
#2.5
print("Number of guess by group:")
print(lines.map(lambda x: (x[3], 1)).reduceByKey(lambda x, y: x+y).collect())

#2.6
print("Number of guess by group in decending order:")
print(lines.map(lambda x: (x[3], 1)).reduceByKey(lambda x, y: x+y).sortByKey(ascending= False).collect())

#2.7
print("How many times President Barack Obama appeared on The Daily Show :",lines.filter(lambda x: x[4] == 'President Barack Obama').count())
```

```
# Question 3

#3.1
raw_data=sc.textFile("/FileStore/tables/q3.txt")
results_line = raw_data.map(lambda line: line.split("\t"))
#3.2
rl_header = results_line.first()
results_line = results_line.filter(lambda row: row != rl_header)
print("Number of entries stored :",results_line.count())
#3.3
print("Number of cars having MPG >20 :",results_line.filter(lambda x: int(x[2]) > 20).count())
#3.4
print("Number of Chevrolet cars that have more than 4 Cylinders :",results_line.filter(lambda x: (x[0] == 'chevrolet') & (int(x[3])>4 )  ).count())
#3.5
def checkCountry(x):
  return (x.lower() == 'american' or x.lower() == 'european')
print('The number of cars manufactured by an American or European company:',results_line.filter(lambda x: checkCountry(x[9]) == True).count())
```
